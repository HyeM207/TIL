## 2. Get 요청하기

```python
import requests
# Get 요청하기
res = requests.get("https://www.naver.com")

res # HTTP 응답아 담김
```

```python
# Header 확인 
res.headers
```

```python
# Body 확인 
res.text
```

## 3. POST 요청하기
POST 실습을 위해 다음 사이트를 이용해본다. [https://webhook.site](https://webhook.site)

사이트에 접속하면 실습을 위한 고유 URL (unique URL)을 주니 해당 URL을 copy하여 아래 실습에 이용한다.

```python
# payload와 함께 POST 요청을 보낸다.
# url은 위에 사이트 접속하여 받은 unique URL을 이용한다.

payload = {"name" : "Hyemin", "age" : 24 }
res = requests.post("https://webhook.site/6b9990c7-3bcd-4ac3-ac10-f5d8a0a644a9", payload)
```

```python
# 응답 코드 확인 
# 응답의 상태코드(status code 확인)

res.status_code
```

200이니 잘 요청된 것을 확인 가능하다. 
요청 시 같이 보낸 payload도 잘 도착했는지 확인하기 위해 위의 사이트에 다시 접속하면 요청과 응답을 확인할 수 있다.
